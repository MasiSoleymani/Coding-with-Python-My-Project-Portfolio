#### For Loop
    * Use for Loop for specifying iteration , which allows code to execute a set of statements once for each item in a list, tuple, set etc. 
    when you use for loop with list, set, tuple dictionary we should use 'in' keyword. When we use for loop with range functoin again we need to use 'in' keyword.

```python
# you can write the same code in while loop or for loop.
#in for loop you use keyword like 'in'
#iterating through a list using a for loop
#list: in list index starts with 0 and length is the length of the number. for the 
#numbers below length is 9.
#add_num = add_num + element means:
# in the second line it says ad_num is 0, and element is 10. So, add_num = add_num + element means,
#0+10= 10. Now add_num is 10, plus element which is 20 becomes 30. 
# element is variable that itereated values through the list.

#First elemetn is 10, second is 20.Elemetn does not indicate index position. 
#iteration will automathically will stops when all element used. So no need to write index line:
#index = index + 1
# the reason why is that value will shifted auto,aticallu starting from first value untill the end of value.

mylist = [10, 20, 30, 40, 50, 60, 70, 80, 90]

add_num = 0

for element in mylist:
    # shift statement
    add_num = add_num + element

print("Sum of array is :", add_num)
```

#### Continue & Break Statements in for 'for loop' and 'while loop'
break function means that if condition is true it stops the loop. Break statement cloese the neaest and closest loop.it breaks out the nearest for loop or while loop.
continue function means that even if the condition is true it continue to the last value.

```python
# if you want simply print all values write:
stocks = ['AAPLE','AMAZON','GOOGLE', None, None,'ORACLE','MICROSOFT']

for value in stocks:
    print(value)
```

```python
# list of company names with None(no value)
# in this code we need just non-non values.
# 'continue' function skips the non and continue to the end of the list/ loop if it is still true but if we use 'break' function
#after printing non, it will stops.
#here I am telling the code when you reach the vlaue 'None' stop, otherwise continue and print the value.
stocks = ['AAPLE','AMAZON','GOOGLE', None, None,'ORACLE','MICROSOFT']
# is and none are value
for value in stocks:
    if (value is None):
        # condition to terminate for loop
        break
    print(value)
```

```python
# list of company names with None(no value)
#here I am telling the code when you reach the vlaue 'None' continue.
stocks = ['AAPLE','AMAZON','GOOGLE', None, None,'ORACLE','MICROSOFT']

for value in stocks:
    if (value is None):
        # condition to continue for loop
        #even if you find None continue. Skip the None, print the value and continue.
        continue
    print(value)
```
